```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/heart.csv")

df.head()
```

```python
df["Sex"].unique()
```

```python
masculino = df[df["Sex"] == "M"]
feminino = df[df["Sex"] == "F"]
```

```python
risco_m = masculino["HeartDisease"].mean() * 100  # porcentagem
risco_f = feminino["HeartDisease"].mean() * 100  # porcentagem

print(f"Risco (Homens): {risco_m:.2f}%")
print(f"Risco (Mulheres): {risco_f:.2f}%")
```

```python
grupos = ["Homens", "Mulheres"]
riscos = [risco_m, risco_f]

plt.figure(figsize=(6, 4))
plt.bar(grupos, riscos, color=["blue", "pink"])
plt.ylabel("Risco de Doença Cardíaca (%)")
plt.title("Comparação do Risco entre Homens e Mulheres")
plt.ylim(0, 100)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
